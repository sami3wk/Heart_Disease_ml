# Type Converting of some columns

```python
def type_converter(df):
    con_list = []
    df_fin = df.columns[0]
    for i in df[df_fin]:
        con_list.append(str(i))
    A[df_fin] = con_list
```

# Data Preprocessing

```python
X = md.preprocessing(X)
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from pandas import DataFrame
le = LabelEncoder()
Y = DataFrame(le.fit_transform(Y),columns=["HeartDisease"])
```

# Split X and Y into trainnig and testing data

```python
from ml_functions import Data_split
ds = Data_split()
xtrain,xtest,ytrain,ytest = ds.train_test(X,Y,32)
```

# Find Best Model for problem

```python
from ml_functions import Model_maker
mm = Model_maker()
```

# a)Logistic linear Classifier

```python
model_logistic = mm.maker_Logistic_model(xtrain,ytrain)
```

```python
from ml_functions import Model_accuracy_measure
mam = Model_accuracy_measure()
lr_tr_acc,lr_ts_acc = mam.classifier_acccuracy(xtrain,ytrain,xtest,ytest,model_logistic)
print("Logistic Classifier Model Train Accuracy",lr_tr_acc)
print("Logistic Classifier Model Test Accuracy",lr_ts_acc)
```

# b) Naive Bayes Classifier

```python
model_nb = mm.maker_naive_bayes_model(xtrain,ytrain)
```

```python
nb_tr_acc,nb_ts_acc = mam.classifier_acccuracy(xtrain,ytrain,xtest,ytest,model_nb)
print("Naive Bayes Classifier Model Train Accuracy",nb_tr_acc)
print("Naive Bayes Classifier Model Test Accuracy",nb_ts_acc)
```

# c)KNN Classifier 

```python
from ml_functions import NeighborsKNN
nknn = NeighborsKNN()
param = nknn.KNN_no_Neighbors_classi(10,xtrain,ytrain)
```

```python
param
```

```python
model_knn = mm.maker_knn_classifier_model(xtrain,ytrain,5)
```

```python
knn_tr_acc,knn_ts_acc = mam.classifier_acccuracy(xtrain,ytrain,xtest,ytest,model_knn)
print("KNN Classifier Model Train Accuracy",knn_tr_acc)
print("KNN Classifier Model Test Accuracy",knn_ts_acc)
```

# d) Decision Tree Classifier with purning

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
dtc = DecisionTreeClassifier(random_state=21)
tg = {"min_samples_split":range(2,6,1),"min_samples_leaf":range(2,6,1),"max_depth":range(2,6,1)}
cv = GridSearchCV(dtc,tg,scoring="neg_mean_absolute_error",cv=4)
cvmodel = cv.fit(xtrain,ytrain)
cvmodel.best_params_
```

```python
model_dtc = mm.maker_dt_classifier_model(xtrain,ytrain,21,4,5,2)
```

```python
dtc_tr_acc,dtc_ts_acc = mam.classifier_acccuracy(xtrain,ytrain,xtest,ytest,model_dtc)
print("Decision Tree Classifier Model Train Accuracy",dtc_tr_acc)
print("Decision Tree Classifier Model Test Accuracy",dtc_ts_acc)
```

# e) Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=21)
tg = {"max_depth":range(2,20,1),"n_estimators":range(2,20,1)}
cv = GridSearchCV(rfc,tg,scoring="neg_mean_absolute_error",cv=4)
cvmodel = cv.fit(xtrain,ytrain)
cvmodel.best_params_
```

```python
model_rfc = mm.maker_rand_for_classi_model(xtrain,ytrain,19,21,14)
```

```python
rfc_tr_acc,rfc_ts_acc = mam.classifier_acccuracy(xtrain,ytrain,xtest,ytest,model_rfc)
print("Random Forest Classifier Model Train Accuracy",rfc_tr_acc)
print("Random Forest Classifier Model Test Accuracy",rfc_ts_acc)
```

# f) Adaboost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier

tg = {"n_estimators":range(2,20,1)}
abcm = AdaBoostClassifier(DecisionTreeClassifier(random_state=21,max_depth=2),random_state=21)
cv = GridSearchCV(abcm,tg,scoring="neg_mean_absolute_error",cv=4)
cvmodel = cv.fit(xtrain,ytrain)
cvmodel.best_params_
```

```python
model_adbc = mm.maker_Adb_classi_model(xtrain,ytrain,21,21,5,2)
```

```python
adbc_tr_acc,adbc_ts_acc = mam.classifier_acccuracy(xtrain,ytrain,xtest,ytest,model_adbc)
print("Adaboost Classifier Model Train Accuracy",adbc_tr_acc)
print("Adaboost Classifier Model Test Accuracy",adbc_ts_acc)
```
